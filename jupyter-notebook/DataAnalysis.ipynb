```python
import pandas as pd
import numpy as np

def create_df(year):
    return pd.read_csv('../Data/ACO.SSP.PUF.Y{}.csv'.format(year), index_col=0)

def add_col_prefixes(df, yy):
    return df.add_prefix('{}_'.format(yy))

PUF_Y2016 = create_df(2016)
PUF_Y2015 = create_df(2015)
PUF_Y2014 = create_df(2014)
PUF_Y2013 = create_df(2013)

#add prefixes to all columns prior to joining dfs
PUF_Y2016 = add_col_prefixes(PUF_Y2016, 16)
PUF_Y2015 = add_col_prefixes(PUF_Y2015, 15)
PUF_Y2014 = add_col_prefixes(PUF_Y2014, 14)
PUF_Y2013 = add_col_prefixes(PUF_Y2013, 13)

#join all tables
PUF_Y15_Y16 = PUF_Y2016.join(PUF_Y2015, how='left')
PUF_Y14_Y16 = PUF_Y15_Y16.join(PUF_Y2014, how='left')
PUF_Y13_Y16 = PUF_Y14_Y16.join(PUF_Y2013, how='left')

PUF_Y13_Y16
```
